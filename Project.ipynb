## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

```python
# Chart - 1 visualization code


# Set the style for seaborn plots
sns.set(style="whitegrid")

# countplot to visualize the distribution of fraud (Class 1) vs. non-fraud (Class 0) transactions
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=df)
plt.title("Distribution of Classes (Fraud vs. Non-Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

##### 1. Why did you pick the specific chart?

The countplot chart was chosen to visualize the distribution of fraud (Class 1) vs. non-fraud (Class 0) transactions because it effectively illustrates the class imbalance in the dataset.

##### 2. What is/are the insight(s) found from the chart?

*  **The countplot shows that the dataset is highly imbalanced, with a significantly higher number of non-fraud transactions (Class 0) compared to fraud transactions (Class 1).**

#### Chart - 2

```python
# Chart - 2 visualization code

# Plot histograms of transaction amounts for fraud and non-fraud transactions
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, label='Non-Fraud', kde=True)
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, label='Fraud', kde=True)
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Count")
plt.legend()
plt.show()
```

##### 1. Why did you pick the specific chart?


Histograms are often used to visualize the distribution of a single variable, making it easy to understand how data is spread across different values. In the context of the provided dataset, histograms were chosen to visualize the distribution of transaction amounts for fraud and non-fraud transactions because they effectively show the spread of amounts and can highlight any differences between the two classes.

##### 2. What is/are the insight(s) found from the chart?

*   **The histogram of transaction amounts for fraud and non-fraud transactions shows differences in the distribution of transaction amounts between the two classes.**

#### Chart - 3

```python
# Chart - 3 visualization code

# Explore the time distribution of transactions
plt.figure(figsize=(10, 6))
sns.histplot(df['Time'], bins=100, kde=True)
plt.title("Distribution of Transaction Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Count")
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

*  **The distribution of transaction times may reveal patterns in the data, such as potential time-based fraud trends.**

#### Chart - 4 - Correlation Heatmap

```python
# Correlation Heatmap visualization code

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Set title
plt.title('Correlation Heatmap')

# Display the plot
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

* **The correlation matrix heatmap can help identify potential relationships between numerical features. Variables with high absolute correlation values may be relevant for modeling.**

#### Chart - 5 - Pair Plot 

```python
# Pair Plot visualization code

sns.pairplot(df[['V1', 'V2', 'V3', 'V4', 'V5', 'Class']], hue='Class', markers=['o', 's'], diag_kind='kde')
plt.title("Pairplot of Selected Features")
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

* **The pairplot allows you to visualize the distribution and relationships between selected features, especially between fraud and non-fraud transactions.**
